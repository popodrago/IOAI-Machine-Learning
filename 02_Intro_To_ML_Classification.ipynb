# Introduction to Classification
### IOAI 2026 - Beginner Track


## 1. Quick Recap

Machine Learning learns patterns from data.

Features (X) → Input  
Labels (y) → Output

In the previous notebook, we predicted numbers.

Now we predict categories.


## 2. What is Classification?

Classification is predicting categories.

Examples:

- Pass or Fail  
- Spam or Not Spam  
- Cat or Dog  
- Fraud or Not Fraud  

The output is not a number.

It is a class.


## 3. Classification vs Regression

Regression → Predicts numbers  
Classification → Predicts categories  

Example:

Regression:
Predict exam score (75, 80, 90)

Classification:
Predict pass (1) or fail (0)


## 4. Binary Classification

Binary classification means two classes.

We use:

0 → Negative class  
1 → Positive class  

Example:

0 = Fail  
1 = Pass


## 5. Create Simple Dataset

We predict whether a student passes based on study hours.

Hours studied → Feature  
Pass (1) / Fail (0) → Label


```python
import numpy as np

# Study hours
X = np.array([[1], [2], [3], [4], [5], [6], [7]])

# Pass (1) or Fail (0)
y = np.array([0, 0, 0, 1, 1, 1, 1])
```

## 6. Train-Test Split

We split data into:

Training set → Teach the model  
Test set → Evaluate the model  

We do not test on training data.


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

## 7. Logistic Regression

Logistic Regression is used for classification.

It predicts probability.

If probability > 0.5 → Class 1  
If probability ≤ 0.5 → Class 0


```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

## 8. Make Predictions


```python
predictions = model.predict(X_test)

print("Predicted classes:", predictions)
print("Actual classes:   ", y_test)
```

## 9. Predict Probabilities


```python
probabilities = model.predict_proba(X_test)

print("Probabilities:")
print(probabilities)
```

Each row shows:

[Probability of class 0, Probability of class 1]


## 10. Accuracy

Accuracy measures how many predictions are correct.

Accuracy = Correct Predictions / Total Predictions


```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
```

Accuracy close to 1 means good performance.

But accuracy alone is not always enough.

We will study better metrics later.
