# Introduction to Machine Learning
### IOAI 2026 - Beginner Track

## 1. What is Intelligence?

Intelligence means:

- Learning from experience  
- Recognizing patterns  
- Solving problems  
- Adapting to new situations  

Example:

If you touch fire and get burned,  
you learn not to touch it again.


## 2. What is Artificial Intelligence?

Artificial Intelligence (AI) is the field of building systems that can perform tasks that normally require human intelligence.

Examples:

- Face recognition  
- Language translation  
- Recommender systems  
- Medical diagnosis  

Inside AI, there is Machine Learning.


## 3. What is Machine Learning?

Machine Learning (ML) is a branch of AI where:

Instead of programming rules manually,  
we allow the computer to learn patterns from data.

Traditional Programming:
Input + Rules → Output

Machine Learning:
Input + Output → Learn Rules


## 4. What is Data?

Data is information.

Examples:

- Numbers  
- Images  
- Text  
- Audio  

In Machine Learning, data has:

Features (X) → Input  
Labels (y) → Correct answer


Example:

Hours Studied → Feature  
Exam Score → Label  

The goal is to learn the relationship between them.


## 5. Types of Machine Learning

1. Supervised Learning  
   We have inputs and correct outputs.

2. Unsupervised Learning  
   We have inputs only.

3. Reinforcement Learning  
   Learning through rewards.


## 6. What is a Model?

A model is a mathematical function.

It maps input to output.

Example:

Score = a × Hours + b

Machine learning finds the best values of a and b.


## 7. Training and Testing

Training set → Used to teach the model.

Test set → Used to evaluate the model.

We test on new data to measure generalization.


## 8. First Machine Learning Example

We will predict exam scores from hours studied.

We use Linear Regression.


```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

Create simple dataset.


```python
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([50, 55, 65, 70, 80])
```

Create model.


```python
model = LinearRegression()
```

Train model.


```python
model.fit(X, y)
```

Make prediction.


```python
prediction = model.predict([[6]])
print("Predicted score:", prediction[0])
```

The model learned a relationship:

Score = a × Hours + b

This is Linear Regression.
